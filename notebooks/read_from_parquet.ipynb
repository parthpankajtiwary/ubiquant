```python
DATA_DIR = '../data/parquet/'
```

##### Read Data #####

```python
%%time
train_df = pd.read_parquet(DATA_DIR+'train_low_mem.parquet')
```

```python
train_df.head(2)
```

```python
agg1 = train_df.groupby('investment_id',as_index=False).agg(avg = ('target', np.mean))
```

```python
agg2 = train_df.groupby('time_id',as_index=False).agg(avg = ('target', np.mean))
```

```python
sns.scatterplot(data=agg2, x="time_id", y="avg")
```

```python
sns.scatterplot(data=agg1, x="investment_id", y="avg")
```
